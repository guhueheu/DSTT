```python
import numpy as np

A = np.reshape(np.arange(36.0), (6,6))
print("Ma trận A (6x6):")
print(A)
print()

I6 = np.identity(6)
print("Ma trận đơn vị I6 (6x6):")
print(I6)
print()

print("Số phần tử của ma trận A:", A.size)
print()

print("Đường chéo của ma trận A:")
print(np.diagonal(A))
print()

A = A + I6
print("Ma trận A sau khi cộng ma trận đơn vị I6:")
print(A)
print()

vecB = np.array([1., 2., 3., 4., 5., 6.])
print("Vector vecB:")
print(vecB)
print()

C = A.dot(vecB)
print("Tích ma trận A x vector vecB:")
print(C)
print()

D = np.array([
    [1., 2., 3., 4., 5., 6.],
    [1., 0., 1., 0., 1., 0.]
])
print("Ma trận D (2x6):")
print(D)
print()

F = np.array([
    [1., 1.],
    [2., 0.],
    [3., 1.],
    [4., 0.],
    [5., 1.],
    [6., 0.]
])
print("Ma trận F (6x2):")
print(F)
print()

G = A.dot(F)
print("Tích ma trận A x F:")
print(G)
print()

try:
    A_inv = np.linalg.inv(A)
    print("Ma trận nghịch đảo của A:")
    print(A_inv)
except np.linalg.LinAlgError:
    print("Ma trận A không khả nghịch (không tồn tại ma trận nghịch đảo)")
print()

print("Tính np.linalg.inv(np.linalg.inv(A)) (nếu A khả nghịch):")
try:
    double_inv = np.linalg.inv(np.linalg.inv(A))
    print("Kết quả:")
    print(double_inv)
    print("Giá trị này phải bằng ma trận A gốc vì nghịch đảo của nghịch đảo ma trận là chính ma trận đó.")
except np.linalg.LinAlgError:
    print("Không thể tính vì A hoặc nghịch đảo của A không khả nghịch.")
print()
```
